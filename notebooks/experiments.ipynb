```python
import numpy as np
import pandas as pd

# Loading California housing pices Regression dataset
from sklearn.datasets import fetch_california_housing

# Loading Breast Cancer Classification dataset
from sklearn.datasets import load_breast_cancer

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score

# Importing the Classification Ensemble learning techniques
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier

# Importing the Regression Ensemble learning techniques
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
```

```python
# 1. Classification Data
cancer = load_breast_cancer()
X_clf, y_clf = cancer.data, cancer.target

# 2. Regression Data
housing = fetch_california_housing()
X_reg, y_reg = housing.data, housing.target

print(f"Classification shape: {X_clf.shape}")
print(f"Regression shape: {X_reg.shape}")
```

```python
# Loading & Splitting the datasets

# Loading Datasets
clf_data = load_breast_cancer()
reg_data = fetch_california_housing()

# Splitting the data for Classification
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(clf_data.data, clf_data.target, test_size=0.2)
# Splitting the data for Regression
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(reg_data.data, reg_data.target, test_size=0.2)
```

```python
# Decision tree for classification
#------------------------------------

# Define the model you want to use
model = DecisionTreeClassifier()

# Train the model
model.fit(X_train_c, y_train_c)

# Predict
y_pred_dt_c = model.predict(X_test_c)

# Calculate the accuracy score
acc_dt_c = accuracy_score(y_test_c, y_pred_dt_c)

# Printing the accuracy
print("Decision Tree Accuracy [Breast Cancer]: ", acc_dt_c) 

```

```python
# Bagging for classification [Random Forest]
#--------------------------------------------

# Defining the model
rf_c = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

# Train the model
rf_c.fit(X_train_c, y_train_c)

# Predict
y_pred_rf_c = rf_c.predict(X_test_c)

# Calculate the accuracy score
acc_rf_c = accuracy_score(y_test_c, y_pred_rf_c)

# Printing the accuracy
print("Random Forest Accuracy [Breast Cancer]: ", acc_rf_c) 
```

```python
# Boosting for classification [Adaboost]
#----------------------------------------

# Defining the model
ada_c = AdaBoostClassifier(
    n_estimators=100,
    random_state=42
)

# Train the model
ada_c.fit(X_train_c, y_train_c)

# Predict
y_pred_ada_c = ada_c.predict(X_test_c)

# Calculate the accuracy score
acc_ada_c = accuracy_score(y_test_c, y_pred_ada_c)

# Printing the accuracy
print("Adaboost Accuracy [Breast Cancer]: ", acc_ada_c) 
```

```python
# Decision tree for Regression
#------------------------------------

# Define the model you want to use
model = DecisionTreeRegressor()

# Train the model
model.fit(X_train_r, y_train_r)

# Predict
y_pred_dt_r = model.predict(X_test_r)

# Calculate the Mean Squared Error
mse_dt = mean_squared_error(y_test_r, y_pred_dt_r)

#R² Score: Tells you how much of the variance in house prices your model explains. 1.0 is a perfect fit, and 0.0 means your model is just guessing the average.
r2_dt = r2_score(y_test_r, y_pred_dt_r)

# Printing the Mean Squared Error
print("Decision Tree (MSE) [Housing]: ", mse_dt) 
print("Decision Tree (r2 score) [Housing]: ", r2_dt) 

```

```python
# Bagging for Regression [Random Forest]
#--------------------------------------------

# Defining the model
rf_r = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

# Train the model
rf_r.fit(X_train_r, y_train_r)

# Predict
y_pred_rf_r = rf_r.predict(X_test_r)

# Calculate the Mean Squared Error
mse_rf = mean_squared_error(y_test_r, y_pred_rf_r)

#R² Score: Tells you how much of the variance in house prices your model explains. 1.0 is a perfect fit, and 0.0 means your model is just guessing the average.
r2_rf = r2_score(y_test_r, y_pred_rf_r)

# Printing the Mean Squared Error
print("Random Forest (MSE) [Housing]: ", mse_rf) 
print("Random Forest (r2 score) [Housing]: ", r2_rf) 
```

```python
# Boosting for Regression [Adaboost]
#----------------------------------------

# Defining the model
ada_r = AdaBoostRegressor(
    n_estimators=100,
    random_state=42
)

# Train the model
ada_r.fit(X_train_r, y_train_r)

# Predict
y_pred_ada_r = ada_r.predict(X_test_r)

# Calculate the Mean Squared Error
mse_ada_r = mean_squared_error(y_test_r, y_pred_rf_r)

#R² Score: Tells you how much of the variance in house prices your model explains. 1.0 is a perfect fit, and 0.0 means your model is just guessing the average.
r2_ada = r2_score(y_test_r, y_pred_ada_r)

# Printing the accuracy
print("Adaboost Accuracy (MSE) [Housing]: ", mse_ada_r) 
print("Adaboost Accuracy (r2 score) [Housing]: ", r2_ada) 
```
